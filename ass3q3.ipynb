```
#Method1
a=[[1, 0], [0, 1]]
b=[[1, 2], [3, 4]]
c=[[],[],[],[]]
d=0
for i in a:
    for j in i:
        for k in b:
            for l in k:
                c[d].append(j*l)
        
        d+=1
        
for i in c:
    print(i)
```
